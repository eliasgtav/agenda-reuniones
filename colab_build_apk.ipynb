```
# 2. Instalar buildozer y cython
!pip install -q buildozer cython==0.29.36
print('✅ Buildozer instalado')
```

```
# 3. Subir el ZIP del proyecto
from google.colab import files
print('Selecciona el archivo agenda.zip que descargaste de tu PC:')
uploaded = files.upload()
```

```
# 4. Extraer el proyecto
import zipfile, os
zip_name = list(uploaded.keys())[0]
with zipfile.ZipFile(zip_name, 'r') as z:
    z.extractall('/content/agenda')
os.chdir('/content/agenda')
!ls -la
print('✅ Proyecto extraído')
```

```
# 5. Compilar APK (tarda 20-40 min la primera vez)
!buildozer -v android debug 2>&1 | tail -50
```

```
# 6. Descargar el APK
from google.colab import files
import glob
apks = glob.glob('bin/*.apk')
if apks:
    print(f'✅ APK generado: {apks[0]}')
    files.download(apks[0])
else:
    print('❌ No se encontró APK — revisa los errores arriba')
```
